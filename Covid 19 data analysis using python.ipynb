# Covid19 Data Analysis using python


### Importing the modules  

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

### Importing covid19 dataset

```python
corona_dataset = pd.read_csv(r"C:/Users/vijay kumar yadav/Desktop/Covid 19 Dataset/covid19_Confirmed_dataset.csv")
corona_dataset.head(10)
```

### checking  the shape of the dataframe

```python
corona_dataset.shape
```

### Removing unnecessary columns

```python
corona_dataset.drop(["Lat","Long"],axis=1,inplace=True) # Axis = 1 is for columns and Axis = 0 is for rows
```

```python
corona_dataset.head(10)
```

### Aggregating the rows by the country

```python
corona_aggregated_dataset = corona_dataset.groupby("Country/Region").sum()
```

```python
corona_aggregated_dataset.head(10)
```

```python
corona_aggregated_dataset.shape
```

### Visualizing data related to a country India

This is a plot of number of peoples infected each day

```python
corona_aggregated_dataset.loc["India"].plot()
```

### Visualizing the comparision of different countries like China,Spain and Italy with India

```python
corona_aggregated_dataset.loc["India"].plot()
corona_aggregated_dataset.loc["China"].plot()
corona_aggregated_dataset.loc["Spain"].plot()
corona_aggregated_dataset.loc["Italy"].plot()
plt.legend()
```

### Calculating a good measure for India

```python
corona_aggregated_dataset.loc["India"][40:95].plot()
```

### Calculating maximum infection rate of India

```python
print("Maximum infection rate in India in one day : ",corona_aggregated_dataset.loc["India"].diff().max(),"citizens")
```

### Finding  maximum infection rate for all of the countries.

```python
countries = list(corona_aggregated_dataset.index)
max_infection_rate = []
for country in countries :
    max_infection_rate.append(corona_aggregated_dataset.loc[country].diff().max())
corona_aggregated_dataset["max_infection_rate"] = max_infection_rate
```

```python
corona_aggregated_dataset.head()
```

```python
corona_max_infection_data = pd.DataFrame(corona_aggregated_dataset["max_infection_rate"])
corona_max_infection_data.head(10)
```

### Importing dataset Worldwide_happiness 
- Importing the WorldHappinessReport.csv dataset
- selecting needed columns for our analysis 
- join the datasets 
- calculate the correlations as the result of our analysis

```python
happiness_report = pd.read_csv(r"C:/Users/vijay kumar yadav/Desktop/Covid 19 Dataset/worldwide_happiness_report.csv") 
happiness_report.head()
```

### Removing unnecessary columns
- Since we are concerned about how covid_19 is affecting happiness worldwide.
- Thus columns such as 'Perceptions of corruption' and 'Generosity' should be removed as they are not dependent on covid_19       situation.
- Also columns such as 'Overall rank' and 'Score' are not concerns of this topic.

```python
unwanted_cols = ["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
happiness_report.drop(unwanted_cols, axis=1, inplace=True)
happiness_report.head()
```

### Changing the index of the dataset 
- Set Country or region	as index.
- It will help to join this dataset with above data set

```python
happiness_report.set_index("Country or region",inplace=True)
happiness_report.head()
```

### Joinning happiness_report and corona_max_infection_data datasets
- Before joinning check the shape of both the datasets

```python
print("Shape of corona_max_infection_data : ",corona_max_infection_data.shape)
print("Shape of happiness_report          : ",happiness_report.shape)
```

```python
# Since both data have different shape we will use inner join
data = corona_max_infection_data.join(happiness_report,how="inner")
data.head()
```

###  Check for correlation matrix
- It is important before creating visualization between two aspects of the dataset.
- It helps us to understand the relation between those parameters.
- We get to know whether the parameters have positive or negative correlation.

```python
data.corr()
```

### plotting GDP vs  max_infection_rate
- This will help us know how GDP of a country affects the infection rate.

```python
x = data["GDP per capita"]
y = data["max_infection_rate"]
# Using seaborn and numpy
sns.scatterplot(x,np.log(y))

# Note here np.log(y) is used to reduce the differnce between x and y values to get a better plot.
# The y values are converted to logarithmic values.
```

```python
# Improvising above plot
sns.regplot(x,np.log(y))
```

### Plotting social support  vs  max_infection_rate
- This will help us know how social support of a country affects the infection rate.

```python
x = data["Social support"]
y = data["max_infection_rate"]
sns.regplot(x,np.log(y))
```

### Plotting healthy life expentancy vs  max_infection_rate
- This will help us know how healthy life expentancy of a country affects the infection rate.

```python
x = data["Healthy life expectancy"]
y = data["max_infection_rate"]
sns.regplot(x,np.log(y))
```
